# **1. Matplotlib Example**
Matplotlib is a versatile Python library for creating static, animated, and interactive visualizations. Here's a simple example to plot a line chart:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a figure and an axes.
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')

# Setting the title and labels
plt.title('Simple Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

Explanation:

This code plots a basic line chart with markers at each data point.

plt.figure(figsize=(8, 5)) sets the size of the figure.

plt.plot() is used to plot the line chart.

plt.title(), plt.xlabel(), and plt.ylabel() are used to add a title and labels to the axes.

# **2. Seaborn Example**
Seaborn is built on top of Matplotlib and integrates closely with pandas data structures. It provides a high-level interface for drawing attractive and informative statistical graphics. Here's an example to create a box plot:

```python
import seaborn as sns
import pandas as pd

# Sample data
data = pd.DataFrame({
    "Group": ["A", "B", "A", "B"],
    "Value": [10, 15, 13, 22]
})

# Create a boxplot
sns.boxplot(x='Group', y='Value', data=data)

# Adding title
plt.title('Box Plot by Group')

# Show the plot
plt.show()
```

# **3. Plotly Example**
Plotly is a library that makes interactive plots. It can create plots that users can interact with, such as zooming in and panning around. Here’s how to create a simple interactive scatter plot:

```python
import plotly.express as px

# Sample data
df = px.data.iris()  # Using iris dataset

# Create a scatter plot
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species',
                 title='Iris Dataset Scatter Plot')

# Show the plot
fig.show()
```
